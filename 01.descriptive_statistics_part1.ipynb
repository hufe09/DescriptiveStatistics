# 数据类型

## Quantitative vs. Categorical  定量与分类

**定量**数据是数值类型，使我们能够进行数学运算（如狗的数量）。

**分类**数据经常用来标记一个群体或一组条目（如狗品种 - 牧羊犬，贵宾犬等）。

## Categorical Ordinal vs. Categorical Nominal 定类数据和定序数据
我们可以将分类数据进一步划分为两种类型：**定序和定类**。

**定序**数据为排名分类，采用排序顺序。（我将这些互动情况评分为非常正面和非常负面，从Very Poor 到 Very Good）。

**定类**数据没有顺序或排名（如男=1,女=2）。

## Continuous vs. Discrete Data 连续与离散

我们可以将定量数据视为**连续**或**离散**。

**连续**数据可以分成更小和更小的单位，并且仍然存在更小的单位。这方面的一个例子是年龄 - 我们可以用年，月，日，小时，秒来衡量年龄的单位，但是仍然有更小的单位可以与年龄相关联。

**离散**数据仅具有可数值。我们互动的狗的数量就是离散数据类型的一个例子。

# Analyzing Quantitative Data 分析数值数据

## 数值数据的四个方面

在分析离散数值数据和连续数值数据时，我们通常会讨论以下四大方面：

1. 集中趋势 `Center`
2. 离散程度 `Spread`
3. 形状`Shape`
4. 异常值`Outliers`

## Analyzing Categorical Data 分析分类数据

通常分析**分类**数据，查看属于每个组的个体的数量或比例。例如，如果查看狗的品种，我们会关心每个品种有多少只狗，或者每种品种的狗的比例是多少。

# Center 集中趋势
Center 有三种衡量标准：
1. `Mean`
2. `Median`
3. `Mode`

## Mean 均值
平均值通常被称为数学中的平均值或**期望值**。我们通过将所有值加在一起，并除以数据集中的值的数量来计算均值。

## Median 中位数

中位数是将数据集一分为二,使 50% 的值更大 而剩下 50% 的值更小。实际计算过程取决于数据集中的数据是偶数个还是奇数个。

### 奇数值的中位数

如果我们有**奇数**个观测值，则**中位数**只是**直接中间**的数字。例如，如果我们有7个观察值，则当我们的数字从最小到最大排序时，中位数是第四个值。如果我们有9个观测值，则中位数是第五个值。
![image](https://raw.githubusercontent.com/hufe09/GitNote-Images/master/Picee/image.8vs7237c1wx.png)

### 偶数值的中位数

如果我们有**偶数**个观测值，则**中位数**是**中间两个值**的**平均值**。例如，如果我们有8个观测值，当我们的数字从最小到最大排序时，中位数为第四个和第五个值的平均值。
![image](https://raw.githubusercontent.com/hufe09/GitNote-Images/master/Picee/image.vyrjqnccg0d.png)
为了计算中位数，我们必须首先对我们的值进行排序。

我们是使用均值还是中位数来描述数据集在很大程度上取决于数据集的**形状**以及是否存在任何**异常值**。

## Mode 众数

**众数**是我们数据集中最常见的值。最常出现的值称为众数。

特定数据集可能有多种众数，或者根本没有众数。
### 没有众数

如果我们的数据集中的所有观察都以相同的频率观察，则没有众数。

如果我们有数据集：1,1,2,2,3,3,4,4

没有众数，因为所有观察值出现次数都相同。

### 多众数

如果两个（或更多）数字共享最大值，则存在多个众数。

如果我们有数据集：1,2,3,3,3,4,5,6,6,6,7,8,9

有两种众数 3 和 6，因为这些值在3倍时共享最大频率，而所有其他值仅出现一次。

# Notation 符号

减号，乘号，除法和等号都有你可能熟悉的数学符号。这些符号中的每一个都取代了数字如何相互作用的想法。在接下来的概念中，您将了解一些与符号相关的其他想法。虽然您不需要使用表示法来完成项目，但它确实具有以下属性：

1. **了解如何正确使用符号使您看起来非常聪明。** 知道如何读写符号就像学习一门新语言一样。用于传达与数学相关的想法的语言。 

2. **它允许您阅读文档，并根据您自己的问题实现一个想法。** 符号用于表示问题是如何一直解决的。一种非常流行的数学算法用于解决世界上一些最困难的问题，称为渐变增强（Gradient Boosting）。它解决问题的方式在这里解释：[https://en.wikipedia.org/wiki/Gradient_boosting](https://en.wikipedia.org/wiki/Gradient_boosting)。如果您真的想了解此算法的工作原理，您需要能够阅读和理解符号。 

3. **它使得难以用语言表达的想法更容易传达。** 有时我们只是没有正确的话语。对于那些情况，我更喜欢使用符号来传达信息。类似于表情符号或模因可以传达一种比文字更好的感觉的方式，符号可以比文字更好地传达一个想法。通常这些想法与数学有关，但我不是在这里扼杀你的创造力。

## 随机变量（Random Variables） 

**随机变量**是某些进程（很多时候……“某些进程”意义比较模糊）的可能值的占位符。如前所述，符号能帮我们简化复杂的意义（通常是单个字母或单个字符）。我们看到随机变量用大写字母表示（$X$、$Y$ 或 $Z$ 是表示随机变量的常用方法）。 

我们可能有随机变量 $X$，它是某人在网站上所花费时间的可能值的占位符。或者随机变量 $Y$，它是某个人是否购买产品的可能值的占位符。 

$X$ 是某人在我们网站花费时间的可能值的'占位符'，它可以是从 0 到无限的任意值。 

## 大写字母与小写字母 

**随机变量**用大写字母表示。每当我们观察到这些随机变量的一个结果，就用相同字母的小写表示。 

- 示例 1 

例如，**某人在我们网站上花费的时间**是一个**随机变量**（我们不确认任意特定访客的结果），我们要用 $X$ 表示它。然后当第一个人访问网站，如果他花费了 5 分钟，现在我们观察到了随机变量的这个结果。我们会用小写字母表示结果，带一个与观察到结果的顺序关联的下标。 

如果有 5 个人访问了我们的网站，第一个花费了 10 分钟，第二个花费了 20 分钟，第三个花了 45 分钟，第四个花了 12 分钟，第五个花了 8 分钟；我们可以用以下方式表示这个问题： 

**X** 是一个人在网站上花费的时间。 

$$
\mathbf{x}_{1}=10, \quad \mathbf{x}_{2}=20 \quad \mathbf{x}_{3}=45 \quad \mathbf{x}_{4}=12 \quad \mathbf{x}_{5}=8
$$

大写 X 与**随机变量**的概念相关联，而此随机变量观察值用小写 x 表示。 

- 示例 2 

我们可以更进一步提问： 

**某人在我们网站花费超过 20 分钟的概率是多少？** 

用符号表达式，我们可以这样写： 

$P(X > 20)?$

这里的$P$代表**概率**，而括号中包含我们要计算概率的语句。由于 $X$ 表示在网站上花费的时间，此符号代表在网站上花费的时间超过 20 分钟的概率。 

在上面的例子中，我们可以注意到，5 个观察值中只有一个超过了 20 分钟。所以，我们说在我们网站上花费超过 20 分钟时间的概率为**五分之一**（45 分钟的那个）或 20%（基于此数据集）。 

- 示例 3 

如果我们问：**一个人在我们网站上花费** **20** **分钟或更多时间的概率是多少？** 我们可以这样表示： 

$P(X≥ 20)?$

我们可以注意到五个人中有两个在网站上花费了 20 分钟或更多时间。所以概率为**五分之二或** **40%**。

## 符号回顾
### 符号和随机变量

作为一个快速回顾，**大写字母**表示**随机变量**。当我们查看特定随机变量的**各个实例**时，我们将这些**实例**标识为**小写字母，**其中下标附加到每个特定的观察。

例如，我们可能有$X$是个人在我们网站上花费的时间。我们的第一位访客到达并在我们的网站上花了10分钟，我们会说 ${X_1}$ 是10分钟。

我们可以将随机变量想象为数据集中的列，而特定值将使用小写字母进行标注。

| **符号**                                                     | **英语**                                                 | **例**           |
| ------------------------------------------------------------ | -------------------------------------------------------- | ---------------- |
| $X$                                                            | 一个随机变量                                             | 花在网站上的时间 |
| ${X_1}$                                                      | 首先观察随机变量X的值                                    | 15分钟           |
| $\sum_{i=1}^{n} x_{i}$            | 求和值从第一次观察开始到最后一次结束                     | 5 + 2 + ... + 3  |
| $\frac{1}{n} \sum_{i=1}^{n} x_{i}$ | 从第一次观察开始到最后一次结束的和值除以观察次数（均值） | （5 + 2 + 3）/ 3 |
| $\overline{x}$                                             | 与上述完全相同 - 我们的数据的平均值。                    | （5 + 2 + 3）/ 3 |

### 均值的表示法

我们通过引入求和的符号来进一步表达我们的符号 \ 和 Σ。使用这个我们能够计算平均值：

$$
\frac{1}{\mathrm{n}} \sum_{\mathrm{i}=1}^{\mathrm{n}} \mathrm{x}_{\mathrm{i}}
$$


# Aggregations 聚合 

**聚合**是一种将多个数字转换为较少数字（通常为一个数字）的方法。 

**求和**是一种常见的聚合方式。用于对值求和的符号是一个希腊字母符号，称为$Sigma \Sigma$

![image](https://raw.githubusercontent.com/hufe09/GitNote-Images/master/Picee/image.l2hak3atyok.png)
- 示例 1 

假设我们在统计访客在我们网站上花费的时间。我们收集了九个人的数据:

$$
\mathbf{x}_{1}=10, \quad \mathbf{x}_{2}=20 \quad \mathbf{x}_{3}=45 \quad \mathbf{x}_{4}=12 \quad \mathbf{x}_{5}=8 \quad \mathbf{x}_{6}=12, \quad \mathbf{x}_{7}=3 \quad \mathbf{x}_{8}=68 \quad \mathbf{x}_{9}=5
$$

如果我们要用以前的表达式对**前三个值**求和，可以写：

$$
\mathbf{x_1} + \mathbf{x_2} + \mathbf{x_3}
$$

使用新表达式，我们可以写：

$$
\sum_{i=1}^{3} x_{i}
$$

注意，我们的表达式从第一个观察值开始 ($i=1$) 到第 3 个值结束（求和符号顶部的数字）。 

所以下面的计算方式结果都是一样的：

$$
\sum_{i=1}^{3} x_{i}=\mathbf{x}_{1}+\mathbf{x}_{2}+\mathbf{x}_{3}=10+20+45=75
$$

- 示例 2 

现在，假设我们要对**最后三个值**求和。

$$
\mathbf{x_7} + \mathbf{x_8} + \mathbf{x_9}
$$

使用新符号，我们可以这样写：

$$
\sum_{i = 7}^9 
$$

注意，我们的符号从第七个观察值 ($i=7$) 开始，到第 9 个值（求和符号顶部的数字）结束。 

## 其他聚合 

$\Sigma$ 符号用于使用求和进行聚合，但是我们可以选择通过其他方式进行聚合。求和是最常见的聚合方式之一。但是，我们可能需要以其他方式进行聚合。如果我们想将所有的值相乘，我们可以使用求积符号 $\Pi$，希腊字母 $\pi$的大写。我们聚合连续值的方式称为积分（微积分中的一种常用技术），它使用以下符号 $\int$，就像一个拉长的 $s$。

## 计算均值的最后步骤 

在均值计算的最后一步，我们引入 **n** 作为我们数据集中值的总数。我们可以将它放在求和符号的顶部，也可以在计算均值时作为分母。 
$$
\frac{1}{n} \sum_{i=1}^{n} x_{i}
$$
我们不写出上面的全部东西，而是通常用 $\bar{x}$ 来表示数据集的均值。不过，与第一个视频中类似，我们可以使用任何变量。因此，我们也可以写 $\bar{y}$，或任何其他字母。 

我们也使用任何其他字母作为索引，而不只是 $i$。我们也可以同样方便地使用$j$、$k$ 或$m$作为每个数据值的索引。
