# 概率  Probability

这节课你学习到概率的一些基本规则。将掷硬币的结果可以为 T 或 H ，分别作为硬币的反面和正面。 

### 然后基本规则为真： 
 
$$
\mathbf{P}(\mathbf{H})=0.5
$$

$$
\mathbf{1}-\mathbf{P}(\mathbf{H})=\mathbf{P}(\text { not } \mathbf{H})=0.5
$$

其中 ${\text{not H}}$ 是除了正面以外的事件。既然只会出现两种可能的结果，我们得到 ${P(\text{not H}) = P(T)} =0.5$. 在随后的概念中，你将看到下列注释： ${\lnot H}$. 

- 通过多次掷硬币，我们得到 n 次正面的概率为 ${P(H)^n}$。因为这些事件是独立的。 

### 我们从中得到下面两个通用规则： 

- 任何事件的概率在 0 和 1 之间，其中包括 0 和 1。 

- 互斥事件的概率为 1 减去某个事件的概率。其他一切可能事件的概率是 1 减去某个事件本身的概率。因此所有可能事件概率的总和等于 1。 

- 如果我们事件是独立的，一系列可能事件的概率是这些事件的乘积。某个事件的概率 AND 下一个事件的概率 AND 下一个事件的概率，即这些事件概率的乘积。 

# 二项分布  The Binomial Distribution

[二项分布](https://en.wikipedia.org/wiki/Probability_mass_function) 帮助我们决定一系列独立的 '掷硬币等事件' 概率。 

与二项分布相关的 概率质量函数 具有以下形式： 

$$
P(X=k) = \frac{n !}{(n-k) ! k !} \cdot p^{k}(1-p)^{(n-k)}
$$

其中 n 是事件数量， k 是 "成功" 的数量，p 是 "成功" 的概率。 

我们现在可以使用这个分布决定下列事件的概率： 

掷硬币 10 次出现 3 次正面的概率。 

掷硬币 10 次出现 8 次以上正面的概率。 

掷硬币 20 次不出现正面的概率。

# 条件概率  Conditional Probability


在这节课中你学习到了条件概率。通常事件并不像掷硬币和骰子一样是独立的。实际上，某个事件的结果依赖于之前的事件。 

例如，得到阳性检验测试结果的概率依赖于你是否具有某种特殊条件。如果具备条件，测试结果就是阳性的。我们通过以下方式用公式表示任意两个事件的条件概率： 

$$
P(A | B)=\frac{P(A \cap B)}{P(B)}
$$

在这个例子中，我们得到下列内容： 

$$
P(\text {positive} | \text {disease})=\frac{P(\text { positive } \cap \text { disease })}{P(\text {disease})}
$$

其中 $|$ 代表 "鉴于"，$\cap$ 代表 "和".

### 机器人传感

步骤细分 

了解解决方案的逐步细分很重要。让我们回顾一下解决方案视频中的内容。 

先验概率 

机器人完全不知道自己在哪里，先验概率如下 

$$
\begin{array}{l}{P(\text { at red })=0.5} \\ {P(\text { at green })=0.5}\end{array}
$$

条件概率 

机器人传感器不是十全十美的。 仅仅因为机器人看见 红色，并 不代表 机器人是在红色格子中。 

$$
\begin{array}{l}{P(\text { see red } | \text { at red })=0.8} \\ {P(\text { see green } | \text { at green })=0.8}\end{array}
$$

后验概率 

从先验概率和后验概率中，我们需要计算机器人看见红色之后的后验概率： 

$$
\begin{array}{l}{\text { 1. } P(\text { at red } | \text { see red })} \\ {\text { 2. } P(\text { at green } | \text { see red })}\end{array}
$$

提示，贝叶斯法则如下所示: 

$$
P(A | B)=\frac{P(B | A) \cdot P(A)}{P(B)}
$$

我们可以将贝叶斯法则中的 A 和B 替换掉，显示为： 

$$
P(\text { at red } | \text { see red })=\frac{P(\text { see rediat red) } \cdot P(\text { at red })}{P(\text { see red })}
$$

现在，我们了解先验概率和条件概率后，可以改写为： 

$$
P(\text { at red } | \text { see red })=\frac{0.8 \cdot 0.5}{P(\text { see red })}
$$

不过我们还不知道一件事！我们看见红色的概率是多少？ 答案是 0.5。我可以用两种方式来证明，一是直观预测，二是数学。 

为什么 P(see red) 是0.5? 

方式1：直观预测 

当然是 0.5！不然呢? 机器人有 50％ 的概率认为自己在红色格子里，有50％的概率认为自己在绿色格子里。当然，它的传感器是不可靠的，但不可靠性是对称的，而不是错误地看到任何一种颜色。 所以无论看到红色的可能性是多少，这也将是看到绿色的可能性。由于这两种颜色是唯一可能的颜色，所以看到每种颜色的概率一定是 50％ ！ 

方式 2: 数学 （全概率法则） 

机器人看到红色有以下两种情况。 

当机器人处于红色格子并且其传感器正常工作时。 

当机器人在绿色格子中，其传感器犯了一个错误。 我只需要把这两个概率加起来就可以得到红色的总概率。 

$$
P(\text { see red })=P(\text { at red }) \cdot P(\text { see red } | \text { at } \operatorname{red})+P(\text { at green }) \cdot P(\text { see red } | \text { at green })
$$

我可以由此得出答案： 

$$
\begin{array}{l}{P(\text { see red })=0.5 \cdot 0.8+0.5 \cdot 0.2} \\ {P(\text { see red })=0.4+0.1} \\ {P(\text { see red })=0.5}\end{array}
$$

$$
P(\text { at red } | \text { see red })=\frac{0.8 \cdot 0.5}{P(\text { see red })} =\frac{0.4}{0.5} =0.8
$$

$$
P(\text { at green } | \text { see red })=\frac{0.2 \cdot 0.5}{P(\text { see red })} =\frac{0.1}{0.5} =0.2
$$
